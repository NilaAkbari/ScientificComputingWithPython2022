#2. Outer product



```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
```

1) Using the function outer in numpy

```python
import numpy as np
print(np.outer(u, v))
```

2) Using a nested for loop or a list comprehension


```python
outpt = []

for i in u:
  uvi = []
  for j in v:
    uvi.append(i*j)

  outpt.append(uvi)

print(outpt)
```

```python
from numpy import newaxis, array

print(u[:, newaxis] * v)
```

#4. Trigonometric functions



```python
import math
nums = np.linspace(0, 2*np.pi, 100)
print(nums)
print("every tenth element is: \n",nums[9::10])
print("reversing the array: \n", nums[99:0:-1])
extracted = []
for i in nums:
  if np.abs(math.sin(i) - math.cos(i)) < 0.1:
    extracted.append(i)

print("the extracted values are: \n", extracted)

sine = np.sin(nums)
cosine = np.cos(nums)

import matplotlib.pyplot as plt

print("the sine and cosine plots: \n")

plt.plot(nums,sine,nums,cosine)
```

#5. Matrices



```python
x = range(1,11) 

mult_table = [[i*j for j in x] for i in x]

mult_table
```

```python
print("The trace is: ", np.trace(mult_table))
```

```python
def anti(m):
  ar = []
  for i in range(len(m)):
    ar.append(m[i][len(m)-i-1])

  return ar
  
print("The anti-diagonal is:", anti(mult_table))
```

```python
print("the diagonal offset by 1 upwards is: \n", np.diag(mult_table, 1))
```

#7. Prime numbers sieve



```python
import math

def fillPrimes(chprime, high):

  ck = [True]*(high+1)

  l = int(math.sqrt(high))
  for i in range(2, l+1):
	  if ck[i]:
		  for j in range(i*i, l+1, i):
			  ck[j] = False
				
  for k in range(2, l+1):
	  if ck[k]:
		  chprime.append(k)
  
  return 

def segmentedSieve(low, high):
	
	chprime = list()
	fillPrimes(chprime, high)
	prime = [True] * (high-low + 1)
	for i in chprime:
		lower = (low//i)	
		if lower <= 1:
			lower = i+i
		elif (low % i) != 0:
			lower = (lower * i) + i
		else:
			lower = lower*i
		for j in range(lower, high+1, i):
			prime[j-low] = False
			
			
	for k in range(low, high + 1):
			if prime[k-low]:
				print(k, end=" ")


low = 2
high = 99
print("Primes in Range %d to %d are"%(low,high))
segmentedSieve(low, high)
```

#8. Diffusion using random walk

```python
from random import choice

walkers = [[choice([-1, 1]) for i in range(200)] for j in range(1000)]
distances = []
for k in range(len(walkers)):
  s = sum(walkers[k])
  distances.append(s)
  print("walker number %d's total distance is %d  and the steps are:" %(k, s), walkers[k])
```

square of each walking path

```python
squared = [j**2 for j in distances]
for k in range(1000):
  print("the square distance for walker %d is %d" %((k+1), squared[k]))
```

mean of squared distances

```python
print("The mean of squared distances is: ", sum(squared)/1000)
```
